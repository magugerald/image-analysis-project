# Image Analysis

In this image analysis project, I focused on enhancing and analyzing digital images using a variety of techniques. The project encompassed computing image statistics and histograms to gain insights into pixel distributions, followed by histogram equalization to improve contrast. Binarization was employed to convert images into black and white by thresholding pixel values. To enhance image quality, noise reduction was achieved through Median Blur and Gaussian Blur, with experimentation across different blur scales. Edge detection was a pivotal aspect, involving Prewitt, Sobel, Laplacian of Gaussian, and Canny edge detectors, revealing prominent edges and contours within images. Overall, the project aimed to harness these techniques to preprocess and extract essential information from images, ultimately contributing to tasks like feature recognition and pattern analysis across various domains.

```python
# Importing Libraries
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
#Importing the image
img = cv2.imread("pic342.jpg")
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

## Image Statistics

```python
# Image Statistics
print ('Min:', np.min(img))
print ('Max:', np.max(img))
print ('Median:', np.median(img))
print ('Mean:', np.mean(img))
print ('Std:', np.std(img))
print ('Var:', np.var(img))
```

## Image Histogram

```python
#Displaying the histogram
plt.hist(img.ravel(), 256, [0,256])
plt.show()

#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

## Histogram Equalization

```python
#spliting the image
R, G, B = cv2.split(img1)

# Creating equalization on the image
R_output = cv2.equalizeHist(R)
G_output = cv2.equalizeHist(G)
B_output = cv2.equalizeHist(B)

# merging the final image
merged_output = cv2.merge((R_output, G_output, B_output))
```

### Plotting histogram equalization

```python
#Plotting
fig = plt.figure(figsize=(12, 7))

fig.add_subplot(2, 2, 1)

# printing original image

plt.imshow(img1)
plt.title("original image")

fig.add_subplot(2, 2, 2)

# printing equalized image
plt.imshow(merged_output)
plt.title("equalized image")

fig.add_subplot(2, 2, 3)

# printing original image
plt.hist(img1.ravel(), 256, [0,256])
plt.title("original histogram")

fig.add_subplot(2, 2, 4)

# printing equalized image
plt.hist(merged_output.ravel(), 256, [0,256])
plt.title("equalized histogram")
plt.show()
```

## Binarizing the image

```python
#Loading the image in grayscale
img0 = cv2.imread("pic342.jpg", cv2.IMREAD_GRAYSCALE)

#Binarizing ( values below 127 moves to 0 i.e black, and everything 127 moves to 255 i.e. white)
_, thresh1 = cv2.threshold(img0, 127, 255, cv2.THRESH_BINARY)

#Plotting the images
plt.subplot(1, 2, 1) 
plt.imshow(img0)
plt.title("original image")

plt.subplot(1, 2, 2) 
plt.imshow(thresh1)
plt.title("Binarized image")
```

## Prewitt and Sobel edge detectors

```python
# Loading the input image
img = cv2.imread('edge.jpg', 0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Defining the Prewitt and Sobel operators
prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

# Applying Prewitt and Sobel operators to the image
edges_prewitt_x = cv2.filter2D(img, -1, prewitt_x)
edges_prewitt_y = cv2.filter2D(img, -1, prewitt_y)
edges_sobel_x = cv2.filter2D(img, -1, sobel_x)
edges_sobel_y = cv2.filter2D(img, -1, sobel_y)

# Combining the x and y gradient images to get the final edge map
edges_prewitt = cv2.addWeighted(edges_prewitt_x, 0.5, edges_prewitt_y, 0.5, 0)
edges_sobel = cv2.addWeighted(edges_sobel_x, 0.5, edges_sobel_y, 0.5, 0)


# Displaying the original image, Prewitt and Sobel edges image
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')
axs[1].imshow(edges_prewitt)
axs[1].set_title('Prewitt Edges')
axs[1].axis('off')
axs[2].imshow(edges_sobel)
axs[2].set_title('Sobel Edges')
axs[2].axis('off')
plt.show()
```

## Laplacian of Gaussian (LoG) edge detector

```python
# Loading the image
img = cv2.imread('clear.jpg', cv2.IMREAD_GRAYSCALE)

# Displaying the Original image
plt.imshow(img)
plt.title("Original Image")

# Seting the range of scales to test
scales = [3, 5, 7, 9]

# Creating a plot with subplots for each scale
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Computing and plotting the LoG for each scale
for i, ax in enumerate(axs.flat):
    if i < len(scales):
        # Compute the LoG using a Gaussian kernel of the current scale
        ksize = int(scales[i] * 3)
        kernel = cv2.getGaussianKernel(ksize, scales[i])
        kernel = kernel * kernel.T
        log = cv2.Laplacian(img, cv2.CV_64F)
        log = cv2.filter2D(log, cv2.CV_64F, kernel)

        # Plot the LoG image
        ax.imshow(log, cmap='gray')
        ax.set_title('Scale = {}'.format(scales[i]))
        ax.axis('off')

# Showing the plot
plt.show()
```

## Laplacian of Gaussian (LOG) and Canny Edge detector

```python
# Loading the image
img = cv2.imread('clear.jpg', cv2.IMREAD_GRAYSCALE)

# Computing the edges using Laplacian of Gaussian (LOG) detector
log = cv2.Laplacian(img, cv2.CV_64F)
log = cv2.GaussianBlur(log, (3, 3), 0) # used to smooth the image
log = np.uint8(np.absolute(log))
log_threshold1 = 50
log_threshold2 = 150
log_edges = cv2.Canny(log, log_threshold1, log_threshold2)

# Computing the edges using Canny edge detector
canny_threshold1 = 50
canny_threshold2 = 150
canny_edges = cv2.Canny(img, canny_threshold1, canny_threshold2)

# Displaying the original image, LOG and Canny edges image
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')
axs[1].imshow(log_edges, cmap='gray')
axs[1].set_title('LOG Edges')
axs[1].axis('off')
axs[2].imshow(canny_edges, cmap='gray')
axs[2].set_title('Canny Edges')
axs[2].axis('off')
plt.show()
```
